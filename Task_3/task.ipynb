```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# -------------------------------
# SQLite connection
# -------------------------------

sqlite_conn = sqlite3.connect(r"C:/Users/patel/Desktop/Intership/Task_3/olist.sqlite")
df = pd.read_sql("SELECT * FROM orders", sqlite_conn)
# sqlite_conn = sqlite3.connect(sqlite_db_path)

# -------------------------------
# PostgreSQL connection
# -------------------------------
pg_user = "postgres"
pg_password = quote_plus("postgres@12")
pg_host = "localhost"
pg_port = "5432"
pg_db = "ECommerce"

pg_engine = create_engine(
    f"postgresql+psycopg2://{pg_user}:{pg_password}@{pg_host}:{pg_port}/{pg_db}"
)

# -------------------------------
# Get all table names from SQLite
# -------------------------------
tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(tables_query, sqlite_conn)

# -------------------------------
# Transfer each table
# -------------------------------
for table_name in tables["name"]:
    print(f"Transferring table: {table_name}")

    df = pd.read_sql(f"SELECT * FROM {table_name}", sqlite_conn)

    df.to_sql(
        table_name,
        pg_engine,
        if_exists="replace",   # use "append" if tables already exist
        index=False
    )

print("✅ Data migration completed successfully!")

# Close connection
sqlite_conn.close()
```
